## Histogram Manipulation  for contrast adjustement  


```python
import cv2
import matplotlib.pyplot as plt
path = r'C:\Users\Root\Desktop\Computer Vision\TP\Images\aero1.jpg'
img = cv2.imread(path, 0)
hist = cv2.calcHist([img],[0], None, [256], [0, 256])
plt.plot(hist)
plt.xlim([0, 256])
plt.ylim([0, max(hist)+1])
plt.show()
```

```python
img = cv2.imread(path,0)
equalized_img = cv2.equalizeHist(img)
hist = cv2.calcHist([img],[0], None, [256], [0, 256])
equalized_hist = cv2.calcHist([equalized_img],[0], None, 
[256], [0, 256])
fig, ax = plt.subplots(2, 2, figsize=(9,7))
ax[0, 0].plot(hist)
ax[0,0].set_title("Histogramme de l’image originale")
ax[0, 1].plot(equalized_hist)
ax[0,1].set_title("Histogram egualisé")
ax[1, 0].imshow(img, cmap="gray")
ax[1, 0].set_title("Image originale")
ax[1, 1].imshow(equalized_img, cmap="gray")
ax[1, 1].set_title("Histogram de l’image egalisée")
plt.show()
```

# Filtering for noise reduction 


```python
import numpy as np
import random
img = cv2.imread(path, 0)
row, col = img.shape
number_of_pixels = random.randint(300, 10000)
for i in range(number_of_pixels):
    y_coord = random.randint(0, row - 1)
    x_coord = random.randint(0, col - 1)
    img[y_coord][x_coord] = 255
number_of_pixels = random.randint(300, 10000)
for i in range(number_of_pixels):
    y_coord = random.randint(0, row - 1)
    x_coord = random.randint(0, col - 1)
    img[y_coord][x_coord] = 0
cv2.imwrite('noisy_image.png', img)
cv2.imshow('Image avec bruit "Salt and Papper"', img)
cv2.waitKey(0)


```

# Average Filter

```python

img = cv2.imread('noisy_image.png')
K = np.ones((5,5),np.float32)/25
filtered_image = cv2.filter2D(img,-1, K)
cv2.imshow ('Averaging_filter.png', filtered_image)
cv2.waitKey(0)
```

```python
img = cv2.imread('noisy_image.png')
filtered_image = cv2.blur(img,(5,5))
cv2.imshow ('Averaging_filterMoy.png', filtered_image)
cv2.waitKey(0)
```

# Gaussian Filter 


```python
img = cv2.imread('noisy_image.png')
filtered_image = cv2.GaussianBlur(img,(5,5),0)
cv2.imshow ('Averaging_filterGauss.png', filtered_image)
cv2.waitKey(0)
```

# Median Filter

```python
img = cv2.imread('noisy_image.png')
filtered_image = cv2.medianBlur(img, 3)
cv2.imshow ('Averaging_filterMed.png', filtered_image)
cv2.waitKey(0)
```

# Bilateral Filter

```python
img = cv2.imread('noisy_image.png')
filtered_image = cv2.bilateralFilter(img, 5,75,75)
cv2.imshow ('Averaging_filterBF.png', filtered_image)
cv2.waitKey(0)
```
